## ğŸ“… Weekly Betting Plan Generator

Generate a full week's betting strategy with bankroll progression.

```python
# Generate a weekly betting plan
print(f"ğŸ“… WEEKLY BETTING PLAN")
print("=" * 25)

# Reset for weekly simulation
weekly_bankroll = BANKROLL
weekly_total_stakes = 0
weekly_expected_profit = 0
days_with_bets = 0

# Generate plan for next 7 days
for i in range(7):
    date = (datetime.now() + timedelta(days=i)).strftime('%Y-%m-%d')
    day_name = (datetime.now() + timedelta(days=i)).strftime('%A')
    
    print(f"\nğŸ“… {day_name} ({date}):")
    
    # Update manager's current bankroll
    manager.current_bankroll = weekly_bankroll
    
    # Generate daily plan (using sample matches for simulation)
    daily_bets = manager.generate_daily_betting_plan(target_date=date)
    
    if daily_bets:
        daily_stakes = sum(bet['bet_size'] for bet in daily_bets)
        daily_expected = sum(bet['expected_value'] for bet in daily_bets)
        
        print(f"   ğŸ¯ Bets: {len(daily_bets)}")
        print(f"   ğŸ’° Total Stakes: ${daily_stakes:.2f}")
        print(f"   ğŸ“ˆ Expected Profit: ${daily_expected:.2f}")
        print(f"   ğŸ’¼ Bankroll After: ${weekly_bankroll - daily_stakes + daily_expected:.2f}")
        
        # Update weekly totals
        weekly_total_stakes += daily_stakes
        weekly_expected_profit += daily_expected
        weekly_bankroll = weekly_bankroll - daily_stakes + daily_expected
        days_with_bets += 1
        
        # Show top bet
        best_bet = max(daily_bets, key=lambda x: x['expected_value'])
        print(f"   ğŸ† Best Bet: {best_bet['match']} - {best_bet['market']} (${best_bet['bet_size']:.2f})")
    else:
        print(f"   âŒ No suitable opportunities")

print(f"\n" + "=" * 50)
print(f"ğŸ“Š WEEKLY SUMMARY")
print(f"=" * 50)
print(f"ğŸ¯ Days with Bets: {days_with_bets}/7")
print(f"ğŸ’° Total Weekly Stakes: ${weekly_total_stakes:.2f}")
print(f"ğŸ“ˆ Expected Weekly Profit: ${weekly_expected_profit:.2f}")
print(f"ğŸ’¼ Expected Bankroll Growth: ${weekly_bankroll - BANKROLL:.2f} ({((weekly_bankroll - BANKROLL)/BANKROLL)*100:.1f}%)")
print(f"ğŸ¦ Projected Final Bankroll: ${weekly_bankroll:.2f}")

if weekly_total_stakes > 0:
    weekly_roi = (weekly_expected_profit / weekly_total_stakes) * 100
    print(f"ğŸ“Š Weekly ROI: {weekly_roi:.1f}%")
    print(f"âš–ï¸  Weekly Risk: {weekly_total_stakes/BANKROLL*100:.1f}% of starting bankroll")
```
